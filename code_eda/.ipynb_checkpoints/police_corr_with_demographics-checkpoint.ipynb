## Formatting and Merging

I examined FIPS values to be sure they were objects and I'db e able to move forward. 

```python
police['FIPS'].unique()
```

In order to use the county demographics data, I created a column with the mean overall police scorecard value for each county by first creating a dictionary of FIPS and mean scores and then mapping each county to its appropriate mean score. 

```python
unique_FIPS = [value for value in police['FIPS'].unique()]
mean_score = []

for FIPS in unique_FIPS:
    mean_score.append(police[police['FIPS'] == FIPS]['calc_overall_score'].mean())
    
dictionary = {unique_FIPS[i]: mean_score[i] for i in range(len(unique_FIPS))}
```

```python
# use FIPS/mean score dictionary to map values to new col in demographics
df['avg_county_police_score'] = df['FIPS'].map(dictionary)
```

I continued to do EDA in this notebook, but wanted a saved version of the csv to run PCA on in a modeling notebook.

```python
df.to_csv('../data/merged_police_demographics.csv', index=False)
```

## Examining Correlation and Quartiles

```python
df.corr()['avg_county_police_score'].sort_values()
```

Since the county health rankings (demographics) data I had could potentially be linked to other demographic data such as race, income, etc, I wanted to look into the quartiles of each of my features to be sure there was a range of values and that I wouldn't end up concluding something ethically inappropriate in my final wrap-up.

```python
# creating quartile columns for my variables

cols_not_using = ['index', 'FIPS', 'State', 'County', 'Water_Violation']

df2 = df.copy()

for col in df.columns:
    if col not in cols_not_using:
        df2[f'{col}_quartiles'] = pd.qcut(df2[col], 4, labels=False)
    else: 
        pass
```

```python
# want to see if there are at least 3 high scoring counties within each quartile
# violent crime rate had the highest correlation, so start there
df2[df2['Violent Crime Rate_quartiles'] == 3]['avg_county_police_score'].sort_values(ascending=False)
```

```python
plt.scatter(df2['Violent Crime Rate_quartiles'], df2['avg_county_police_score'], alpha=0.3)
plt.xlabel('Quartiles of Violent Crime Rates - Low to High', size=14)
plt.ylabel('Average County Police Score', size=14)
plt.title('Spread of Police Scorecard Values by Violent Crime Rate Quartiles');
```

**Interpretation:** From my correlation analysis, I found that Violent Crime Rate had the highest correlation with Police Scorecard values (see above correlation, it was around -0.25). I wanted to be sure that each quartile of Violent Crime Rates had high scoring counties. The graph shows the similarity in the spread of average police scores for each quartile of Violent Crime Rate. 

## Initial Model

Below I created an inital model before any preprocessing (like Standard Scaling) to get a baseline of my clusters. 

```python
cols_not_using = ['index', 'FIPS', 'State', 'County', 'avg_county_police_score']

features = [col for col in df.columns if col not in cols_not_using]
X = df[features]

model = KMeans(n_clusters = 7, random_state=42)
model.fit(X)
X = pd.DataFrame(X, columns=features)
```

```python
labels = model.labels_
centroids = model.cluster_centers_

X['cluster'] = labels 
```

```python
score = silhouette_score(X, labels, metric='euclidean')
score
```

```python
X['cluster'].value_counts(normalize=True)
```

Upon examining my initial model, if I had only looked at the silhouette score, it may have appeared to be a good model. However, more than 82% of the data ended up in only two clusters.
