## Manipulating DataFrame

I need to add a column in my DataFrame for use later in my function (not a feature in my clustering model). This column will be the average Police Scorecard (`calc_overall_score`) value for the county.

```python
unique_FIPS = [value for value in police['FIPS'].unique()]
mean_score = []

for FIPS in unique_FIPS:
    mean_score.append(police[police['FIPS'] == FIPS]['calc_overall_score'].mean())
    
dictionary = {unique_FIPS[i]: mean_score[i] for i in range(len(unique_FIPS))}

# use FIPS/mean score dictionary to map values to new col in demographics
df['avg_county_police_score'] = df['FIPS'].map(dictionary)
```

## Principal Component Analysis

I chose to use 2 components in my PCA because there was a big drop in the explained variance ratio after 2 components. Also, I am able to visualize my components in 2-dimensional space.

```python
df[df['# Driving Deaths'].isnull()]
```

```python
# creating feature matrix
not_using = ['index', 'FIPS', 'State', 'County', 'avg_county_police_score']
features = [col for col in df.columns if col not in not_using]

X = df[features]

# scale data for PCA
X_sc = StandardScaler().fit_transform(X)

# initialize PCA with 2 components and fit on the SS data
pca = PCA(n_components=2)
pca.fit(X_sc)
```

Examining the Explained Variance Ratio of each Principal Component

```python
print('Exp Var: ', pca.explained_variance_ratio_)
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(range(1,3), pca.explained_variance_ratio_, lw=2)
ax.scatter(range(1, 3), pca.explained_variance_ratio_, s=100)
ax.set_title('explained variance of components')
ax.set_xlabel('principal component')
ax.set_ylabel('explained variance');
```

I analyzed the weights of each feature to pull out themes from each principal component. I found that the first component weighed heavily on the overall occurrences of different categories. There were no features that represented percentages of the overall population. The top weighted features were population, education, and family related. Ref: I sourced the code for much of the PCA code written here from the PCA General Assembly lecture. 

```python
# first PC/eigenvector w/ ExpVar ~ .322
pca_component_1_eigenvec = pca.components_[0]

# pca_component is the eigenvector describing a new axis
pca_1 = []
for weight, feature in zip(pca_component_1_eigenvec, 
                         X.columns):
    pca_1.append((weight, feature))
sorted(pca_1)
```

In comparison, the second principal component contained only percentage values from the populations. The focus appeared to rely on education and healthcare/healthcare access with a slight emphasis on family demographics. 

```python
# second PC/eigenvector w/ ExpVar ~ .21
pca_component_2_eigenvec = pca.components_[1]

# pca_component is the eigenvector describing a new axis
pca_2 = []
for weight, feature in zip(pca_component_2_eigenvec, 
                         X.columns):
    pca_2.append((weight, feature))
sorted(pca_2)
```

## KMeans Clustering Model

Creating a new DataFrame with the PCA transformed feature matrix.

```python
# transfer scaled data to the new PCA space
X_reduced = pca.transform(X_sc)
pca_df = pd.DataFrame(X_reduced, index=X.index, columns=['PC1', 'PC2'])
pca_df.shape, X_sc.shape
```

This Dataframe represents our new features after having been transformed using the weights of Principal Component 1 and Principal Component 2. 

```python
pca_df.head()
```

Instantiating my clustering model and creating cluster label columns in my PCA dataframe and my original dataframe. 

```python
model = KMeans(n_clusters = 7, random_state=42)
model.fit(pca_df)

labels = model.labels_
pca_df['cluster'] = labels
df['cluster'] = labels
```

Analyzing cluster value counts. Several of my initial models created incredibly unbalanced clusters. For example, I had one model in which 82% of my data was contained in only two clusters (of 7). PCA improved the split as well as my sihouette score tremendously. (My initial silhouette scores were under 0.3). 

```python
pca_df['cluster'].value_counts()
```

```python
score = silhouette_score(pca_df, labels, metric='euclidean')
score
```

## Visualizing Clusters

Code for the visualization adapted from [source](https://openclassrooms.com/en/courses/5869986-perform-an-exploratory-data-analysis/6177861-analyze-the-results-of-a-k-means-clustering).<br>

PCA allowed a much more clear separation of my clusters. (Hence the increase in my silhouette score). 


```python
# code taken from PCA lecture
fig, ax = plt.subplots(figsize=(10,8))

colors = ['#fde725', '#90d743', '#35b779', '#21918c', '#31688e', '#443983', '#440154']
cluster = [0, 1, 2, 3, 4, 5, 6]

for i, col, cluster in zip(range(8), colors, cluster):
    sub = pca_df[pca_df['cluster']==i]
    ax.scatter(sub.iloc[:,0], sub.iloc[:,1], s=150, 
               c=col, label=cluster, alpha=0.3)
    
# plot cluster centers
ax.scatter(x=model.cluster_centers_[:, 0], 
           y=model.cluster_centers_[:, 1], 
           marker='x', s=150, 
           linewidths=3, 
           color='black', 
           zorder=10)
    
ax.set_xlabel('principal component 1')
ax.set_ylabel('principal component 2')
ax.set_title('Projection of Points (on pc1 and pc2)', size=16)
ax.legend();
```

A parallel coordinates plot allows us to see how individual data points sit across ALL variables. By looking at how the values for each variable compare across all clusters, we can get a sense of what each cluster represents. Since the data is so dense, you can see that it's hard to look at differences in each cluster with the lines are close to the mean (standard scaled data). I found the initial code from this [source](https://openclassrooms.com/en/courses/5869986-perform-an-exploratory-data-analysis/6177861-analyze-the-results-of-a-k-means-clustering)

## Function Creation

I wanted a simpler dataframe from which to pull my top scoring counties for each cluster, so I created that here. 

```python
# creating simplified dataframe of top scoring counties in each cluster
indeces = []
for cluster in df['cluster'].unique():
    for i in range(3):
        indeces.append(list(df[df['cluster'] == cluster]\
                            .sort_values('avg_county_police_score', ascending=False)\
                            .head(3).index.values)[i])
# creating dataframe with top scoring counties for each cluster
top_counties = df.loc[indeces, ['State', 'County', 'avg_county_police_score', 'cluster']]
```

```python
# writing function to return highest scoring counties for each cluster 
def top_counties_fn(State, County):
    try:
        lookup = df[(df['State'] == State) & (df['County'] == County)]
        cluster = lookup['cluster'].values[0]
        if lookup.loc[[(lookup.index.values)[0]], ['avg_county_police_score']].isnull().values[0][0] == False:
            print(f'Your current County Average Police Score is {round(lookup.avg_county_police_score.values[0], 2)}')
            print('-------------------')
            print(f'The top 3 counties in your cluster and their scores:')
            top_values = top_counties[top_counties['cluster'] == cluster]
            for i in range(3):
                print(f'Location: {top_values.County.values[i]}, {top_values.State.values[i]} with score of: {round(top_values.avg_county_police_score.values[i], 2)}')

        else: 
            print(f'Your County has no scores currently.')
            print('-------------------')
            print(f'The top 3 counties in your cluster and their scores:')
            top_values = top_counties[top_counties['cluster'] == cluster]
            for i in range(3):
                print(f'Location: {top_values.County.values[i]}, {top_values.State.values[i]} with score of: {round(top_values.avg_county_police_score.values[i], 2)}')
    
    except:
        print('Check your spelling. It is possible we have no data for that county.')

    print('-------------------')
    return 'Consider visiting https://policescorecard.org/ for more information'
```

```python
# testing function
top_counties_fn('Indiana', 'Spen')
```

```python
top_counties_fn('Colorado', 'Denver')
```

**Interpretation:** When supplied with a state and county, the function above provides a Police Scorecard Average for that County, if it exists. It also returns the top three counties in the same cluster. This is meant to provide information to county policy makers. They can use this information as a jumping off point for research regarding what types of policies those high-scoring counties have enacted, as county/city change that has worked in a county similar to theirs may be a better case study for their own change action steps.
