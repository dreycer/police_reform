## Cleaning

```python
# no duplicates
df[df.duplicated()]
```

```python
# limiting to police departments instead of sheriff's depts
df = df[df['agency_type'] == 'police-department']
```

```python
df.head()
```

```python
# the only columns I want are the agency, location, and overall policing score
cols_to_drop = [col for col in df.columns if col not in \
                ['location_name', 'state', 'fips_county_code',
                 'fips_state_code', 'calc_overall_score']]

df = df.drop(columns=cols_to_drop)
```

```python
df.shape
```

```python
df.dtypes
```

The Overall Score column needed to have the % sign removed and needed to be cast as an integer. 

```python
df['calc_overall_score'] = [int(value.replace('%', '')) for value in df['calc_overall_score']]
```

County FIPS codes are actually a combination of the state FIPS code and the County code. I need to have a single code in my table in order to merge it later with my county health record (demographics) data. There were also 4 counties missing a code, so I looked them up and appropriately imputed them.  

```python
df[df['fips_county_code'].isnull()]
```

```python
# https://www.nrcs.usda.gov/wps/portal/nrcs/detail/national/home/?cid=nrcs143_013697
# looked up county for each missing value and imputed appropriate values
df.loc[345, 'fips_county_code'] = 121
df.loc[673, 'fips_county_code'] = 67
df.loc[713, 'fips_county_code'] = 89
df.loc[2166, 'fips_county_code'] = 117
```

```python
df.dtypes
```

I needed my state and county fips codes to be the same data type so I could put them together. 

```python
df['fips_county_code'] = df['fips_county_code'].astype(int)
```

```python
# function to concatenate two numbers
# source: https://stackoverflow.com/questions/12838549/how-to-concatenate-two-integers-in-python
def concat(a, b):
    return (f"{a}{b}")
```

```python
state_2_digits = []
county_3_digits = []
for state, county in list(zip(df['fips_state_code'], df['fips_county_code'])):
    if (len(str(state))) == 2:
        state_2_digits.append(str(state))
    else:
        state_2_digits.append(concat(0, state))

for state, county in list(zip(df['fips_state_code'], df['fips_county_code'])):
    if len(str(county)) == 3:
        county_3_digits.append(str(county))
    elif len(str(county)) == 2:
        county_3_digits.append(concat(0, county))
    else:
        county_3_digits.append(concat('00', county))
        
df['FIPS'] = [concat(state, county) for state, county in list(zip(state_2_digits, county_3_digits))]
df.drop(columns=['fips_state_code', 'fips_county_code'], inplace=True)
```
