## Cleaning

I need all 5 digit FIPS codes, so I will fix that. In our data, the values are given as an integer, which cuts off the leading '0' value. I need to insert that value and will do that with a function and a for loop.

```python
# function to concatenate two numbers
# source: https://stackoverflow.com/questions/12838549/how-to-concatenate-two-integers-in-python
def concat(a, b):
    return (f"{a}{b}")
```

```python
county_5_digits = []
for code in df['FIPS']:
    if (len(str(code))) == 5:
        county_5_digits.append(str(code))
    else:
        county_5_digits.append(concat('0', code))

df['FIPS'] = [code for code in county_5_digits]        
```

Each state's excel file began with a row that was the aggregate for that state. I removed those to put them into their own state specific dataframe while leaving the county data alone.

```python
# creating states dataframe
states = df[df['County'].isnull()]
```

```python
# removing those rows from the original dataframe
df = df[df['County'].notna()]
```

```python
# identifying what percentage of each column's values are null
df.isnull().mean().sort_values(ascending=False)
```

I need to drop rows that have too many missing values. I am using an upper limit to start of 10% of data missing. I will likely remove more later, but need a starting place. 

```python
cols_to_drop = [col for col in df.columns if df[col].isnull().mean() > 0.1]
df.drop(columns=cols_to_drop, inplace=True)
```

```python
df.isnull().mean().sort_values(ascending=False)
```

Removing any columns with 'Z-Score' in label, as Z scores can be calculated for the overall dataset using StandardScaler if needed and will likely be a part of my preprocessing before I model. The current Z-scores only apply to individual states, not the country as a whole. Same is true for Confidence Intervals and Percentiles.

```python
# dropping columns that are Z-Score values
df.drop(df.columns[df.columns.str.contains('Z-Score')], axis=1, inplace=True)

# dropping columns that are confidence interval values
df.drop(df.columns[df.columns.str.contains('95%')], axis=1, inplace=True)

# dropping columns that are percentile values
df.drop(df.columns[df.columns.str.contains('Percentile')], axis=1, inplace=True)
```

Upon realizing that there was only one null value for a few of the columns, I identified a single county in Hawaii that contained 27 null values (in a single row!). The county itself is the smallest in the country, with a population of only 66 people. I chose to drop this county from the dataset based on this information and the liklihood that the data would be usable or trustworthy with so many imputations and so little information on the population itself. 

[Source](https://en.wikipedia.org/wiki/Kalawao_County,_Hawaii) 

```python
df.drop(df[df['% Uninsured'].isnull()].index, inplace=True)
```

In researching the states that had null values for the `Presence of Water Violation`, I found that it was only missing for Hawaii and Virginia. I found information about violations from the EWG and was able to impute. I chose to dummy the column at the same time, where 1 represents the existence of a water violation in the county and 0 represents no water violations in the county.

[Source](https://www.ewg.org/tapwater/state.php?stab=HI)

```python
VA_and_HI_with_violations = ['Suffolk City', 'Petersburg City', 'Chesapeake City', 
                      'Salem City', 'Richmond City', 'Honolulu', 'Maui', 'Kauai']

df['Water_Violation'] = [0 if value not in VA_and_HI_with_violations else 1 for value in df['County']]
df.drop(columns='Presence of Water Violation', inplace=True)
```
